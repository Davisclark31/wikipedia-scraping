```python
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text,"html.parser")
```

```python
table = soup.find_all("table")
```

```python
table[0].find_all("th") 
```

```python
table[0].find_all("td")
```

```python
temp = table[0].find_all("td")
```

```python
data = []
j = 1
subset = []
for i in temp:
    if j%6 == 1 and j != 1:
        data.append(subset)
        subset = []
    subset.append(i.text)
    j += 1
```

```python
col = ["Rank", "Country", "Population", "% of world population", "Date", "Source"]
```

```python
ndf = pd.DataFrame(data, columns = col)
```
